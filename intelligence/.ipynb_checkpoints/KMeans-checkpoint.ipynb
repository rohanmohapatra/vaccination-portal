```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('patient.csv')
df.head()
df_copy = df.copy()
train = df_copy.sample(frac=0.75, random_state = 0)
test = df_copy.sample(frac=0.25, random_state=0)
print(train.head(10))
print(test.head(10))
```

```python
labelEncoder = LabelEncoder()
"""labelEncoder.fit(df['patient_id'])
df['patient_id'] = labelEncoder.transform(df['patient_id'])"""
labelEncoder.fit(df_copy['patient_id'])

df_copy['patient_id'] = labelEncoder.transform(df_copy['patient_id'])
#test['patient_id'] = labelEncoder.transform(df_copy['patient_id'])
```

```python
df_copy
df_copy = df_copy.drop(['Unnamed: 0'], axis=1)
df_copy
```

```python
train = df_copy.sample(frac=0.75, random_state = 0)
test = df_copy.sample(frac=0.25, random_state=0)
print(train.head(10))
print(test.head(10))
```

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit(train)
```

```python
import numpy as np
for i,j in test.iterrows():
    print(kmeans.predict(np.array(j).reshape([-1,21])))
```

```python
results = kmeans.predict(test)
results
```

```python
test.insert(21, "Cluster", results, True)
#df.insert(2, "Age", [21, 23, 24, 21], True) 
test
```
